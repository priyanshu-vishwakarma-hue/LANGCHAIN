##### Sentence Transformers on Hugging Face

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings=HuggingFaceEmbeddings(model_name= "all-MiniLM-L6-v2") # this model is of sentence_transformers and it is installed in venv by requirements.txt
embeddings
```

```python
text="this is a test document"

query_result=embeddings.embed_query(text)

print(len(query_result))
query_result
```

```python
query_result[0]
```

```python
doc_result=embeddings.embed_documents([text,"the second text"])

print(len(doc_result))
doc_result
```
