##### Ollama supports embedding model making it possible to build retrieval augmented generation    RAG model means  application that combine text prompts with existing documents or other data

```python
from langchain_ollama import  OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="phi4-mini:latest")  #by default it uses llama2
)
```

```python
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of greek word",
        "beta is the second letter of greek word"
    ]
)

r1
```

```python
print(len(r1[0])) # dimention of embedding
r1[0]
```

```python
embeddings.embed_query("Alpha is the first letter of greek word")
```
